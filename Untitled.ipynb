## Analysing 4-W cars available around the world.

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df=pd.read_csv('ElectricCarData_Clean.csv')
df.head(2)
```

```python
df2=df[['AccelSec', 'TopSpeed_KmH', 'Range_Km','Efficiency_WhKm', 'FastCharge_KmH', 'RapidCharge', 'PowerTrain', 'BodyStyle', 'Segment', 'Seats', 'PriceEuro']]
```

```python
df2.head(1)
```

In India, 24.04 INR will allow you to buy the same things you'd buy with 1.00 EUR in Germany(Purchasing Power Parity).

```python
df2.rename(columns = {'PriceEuro':'Price(Rs.)'}, inplace = True)
```

```python
df2['Price(Rs.)']=df2['Price(Rs.)'].apply(lambda x:x*24.04)
```

```python
df2=pd.get_dummies(df2,columns=['RapidCharge', 'PowerTrain', 'BodyStyle', 'Segment'],drop_first=True)
```

```python
df2.drop([57,68,77,82,91],inplace=True)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=6)
```

```python
x_pca=pca.fit_transform(df2)
```

```python
pca.explained_variance_ratio_*100
```

First priciple component captures almost all the information.

```python
PC=pd.DataFrame(x_pca,columns=['PC1','PC2','PC3','PC4','PC5','PC6'])
```

```python
PC_composition=pd.DataFrame(pca.components_,index=['PC1','PC2','PC3','PC4','PC5','PC6',],columns=df2.columns)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(PC_composition)
```

```python
def biplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n= coeff.shape[0]
    scalex = 1.5/(xs.max()-xs.min())
    scaley = 1.5/(ys.max()-ys.min())
    fig,ax = plt.subplots(figsize=(12,10))
    
    sns.scatterplot(xs*scalex,ys*scaley)
    
    #for i ,txt in enumerate(names):
        #ax.annotate(txt,(xs[i]*scalex,ys[i]*scaley))
        
    for i in range(n):
        ax.arrow(0,0,coeff[i,0],coeff[i,1],color='r',alpha=0.5)
        if labels is None:
            ax.text(coeff[i,0]*1.15,coeff[i,1]*1.15,"Var"+str(i+1),color='g',ha='center',va='center')
        else:
            ax.text(coeff[i,0]*1.15,coeff[i,1]*1.15,labels[i],color='g',ha='center',va='center')
    ax.set_xlim(-1,1)
    ax.set_ylim(-1,1)
    ax.set_xlabel("PC{}".format(1))
    ax.set_ylabel("PC{}".format(2))
    
```

```python
biplot(x_pca[:,0:2],np.transpose(pca.components_[0:2,:]),labels=df2.columns)
```

It can be seen that for a mid-range price , a car with good features (fastcharge,Range and speed) can be easlily designed. After a price threshold , the features do not necessarily scale.

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,10),metric='silhouette',locate_elbow=False)
visualizer.fit(df2)        
visualizer.show()   
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(15, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(df2, method='ward'))
```

#### From all the above analysis, we can conclude that 4 clusters seems like a good fit for the data.
